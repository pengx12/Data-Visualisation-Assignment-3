```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('whc-sites-2019-2.csv')
gdpdata = pd.read_csv('maddison-data-gdp-per-capita-in-2011us.csv')
traveldata = pd.read_csv('international-tourism-number-of-arrivals.csv')
```

```python
data=data[['unique_number','name_en','date_inscribed','longitude','latitude','category','category_short','states_name_en','region_en']]
```

```python
data.head()
```

```python
gdpdata['date_inscribed']=gdpdata['Year']
gdpdata['states_name_en']=gdpdata['Entity']
gdpdata['gdp']=gdpdata['GDP per capita']
gdpdata=gdpdata[['states_name_en','date_inscribed','gdp']]
```

```python
traveldata['date_inscribed']=traveldata['Year']
traveldata['states_name_en']=traveldata['Entity']
traveldata=traveldata[['states_name_en','date_inscribed','count']]
```

```python
newdata=pd.merge(data, gdpdata,on=['date_inscribed','states_name_en'], left_index=True, how='left')
newdata=pd.merge(newdata, traveldata,on=['date_inscribed','states_name_en'], left_index=True, how='left')
newdata['date_inscribed'] = newdata['date_inscribed'].astype('int')
newdata.sort_values(by=["states_name_en", 'date_inscribed'], ascending=True,inplace=True)
newdata=newdata.fillna(method='pad',axis=0)
newdata=newdata.fillna(0)
newdata.sort_values("date_inscribed",inplace=True)
newdata.head()
```

```python
newdata
```

```python
newdata['count'].count()
```

```python
df = newdata.groupby(by=['category']).size()
df = df.to_frame() 
newdata=pd.merge(newdata, df,on=['category'], left_index=True, how='left')
newdata
newdata=newdata.rename(columns={0:'countbycategory'})
```

```python
df = newdata.groupby(by=['states_name_en']).size()
df = df.to_frame() 
newdata=pd.merge(newdata, df,on=['states_name_en'], left_index=True, how='left')
newdata=newdata.rename(columns={0:'countbystate'})
newdata.head()
```

```python
newdata=newdata.rename(columns={'0_x':'countbycategory'})
newdata=newdata.rename(columns={'0_y':'countbynation'})
newdata.head()
```

```python
df = newdata.groupby(by=['region_en']).size()
df = df.to_frame() 
newdata=pd.merge(newdata, df,on=['region_en'], left_index=True, how='left')
newdata=newdata.rename(columns={0:'countbyregion'})
newdata.head()
```

```python
df = newdata.groupby(by=['date_inscribed']).size()
df = df.to_frame() 
df[0]=df[0].groupby(['date_inscribed']).cumsum()
newdata=pd.merge(newdata, df,on=['date_inscribed'], left_index=True, how='left')
newdata=newdata.rename(columns={0:'countbydate'})
newdata.head()
```

```python
df = newdata.groupby(by=['date_inscribed','region_en']).size()
df = df.to_frame() 
df[0]=df[0].groupby(['region_en']).cumsum()

newdata=pd.merge(newdata, df,on=['date_inscribed','region_en'], left_index=True, how='left')
newdata=newdata.rename(columns={0:'countbyregiondate'})
newdata.head()
```

```python
df = newdata.groupby(by=['date_inscribed','states_name_en']).size()
df = df.to_frame() 
df[0]=df[0].groupby(['states_name_en']).cumsum()
newdata=pd.merge(newdata, df,on=['date_inscribed','states_name_en'], left_index=True, how='left')
newdata=newdata.rename(columns={0:'countbystatedate'})
newdata.head()
```

```python
df = newdata.groupby(by=['date_inscribed','category']).size()
df = df.to_frame() 
df[0]=df[0].groupby(['category']).cumsum()
newdata=pd.merge(newdata, df,on=['date_inscribed','category'], left_index=True, how='left')
newdata=newdata.rename(columns={0:'countbycatdate'})
newdata.head()
```

```python
newdata.to_csv('preprocessed.csv')
```
